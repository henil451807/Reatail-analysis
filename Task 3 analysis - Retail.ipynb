# Now Starting a Data Analysis

```python
# Exploratory Data Analysis
plt.figure(figsize=(16,9))
sns.kdeplot(data['Sales'],color='Blue',label='Sales',shade=True,bw=25)
sns.kdeplot(data['Profit'],color='Green',label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend()
plt.show()
```

```python
# Numerical data
numerical_data=data.select_dtypes(exclude=[object])          
print(numerical_data.shape)
```

```python
# Finding the correlation of numerical data
corr_matrix=numerical_data.corr()
corr_matrix
```

```python
# Visualing the correlation matrix using heatmap
plt.figure(figsize=(16,9))
sns.heatmap(data.corr(),cmap='rocket_r',annot=True)
plt.show()
```

# now plot graphs for Data Visualization

```python
# Plotting Graph
data.hist(bins=50 ,figsize=(20,20))
plt.show()
```

# Pairplot on the basis of Shipping Mode

```python
# Based on Shipping Mode
sns.pairplot(data,hue='Ship Mode')
plt.show()
```

```python
#count value on shipmode
data['Ship Mode'].value_counts()
```

```python
sns.countplot(x='Ship Mode',data=data,palette='tab10')
plt.title('Ship Mode',fontsize=30)
```

# Shipping Mode ( Sales, Profit , Discount) Analysis

```python
data['Ship Mode'].value_counts()
```

```python
df_shipmode = data.groupby(['Ship Mode'])[['Sales', 'Discount', 'Profit']].mean()
```

```python
#1. sales 2. Discount 3. Profit
df_shipmode.plot.pie(subplots=True,
                     figsize=(18, 20), 
                     autopct='%1.1f%%', 
                     labels = df_shipmode.index)
plt.show()
```

# Sub-Category  [Sales, Profit and Discount] Analysis

```python
df_sub_category = data.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub_category.head(10)
```

```python
plt.figure(figsize = (16,16))
plt.pie(df_sub_category['Sales'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 30)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

# Based on SubCategory

```python
plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Discount'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis', fontsize = 30)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

# Based on Profit
